## Membuat Kolom / Menambah Kolom baru

```python
dfMovie = pd.read_csv('/content/drive/MyDrive/DTSProA/imdb.csv')
```

```python
# Menambahkan kolom baru

dfMovie['Nilai'] = dfMovie['Rating'] * 10
```

```python
# Menambahkan kolom baru dengan variable

durasi = (dfMovie['Runtime (Minutes)']/60).round(2)

dfMovie['Durasi'] = durasi
dfMovie
```

```python
# Menambah kolom dengan insert() dan menentukan index kolom

dfMovie.insert(8,'Durasi_Jam', durasi, True)
dfMovie
```

```python
# Menambah kolom dengan assign, dapat digunakan untuk lebih dari satu kolom

nilai = dfMovie['Rating'] + dfMovie['Metascore'] / 2 
pemilih = (dfMovie['Votes'] / 1000).round(2)

dfMovie2 = dfMovie.assign(Durasi_Jam = durasi, Nilai_Rata = nilai, Jumlah_Voting = pemilih)
dfMovie2
```

```python
# Multiple Manipulation

film_vote = dfMovie2[(dfMovie2['Jumlah_Voting'] >= 500) & (dfMovie2['Nilai_Rata'] > 40)]
film_vote_sort = film_vote.sort_values('Durasi_Jam', ascending = False)

film_vote_sort[['Title', 'Year' ,'Durasi_Jam', 'Nilai_Rata', 'Jumlah_Voting']]
```
