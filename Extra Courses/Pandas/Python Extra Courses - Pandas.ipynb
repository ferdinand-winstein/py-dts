A pivot table is a table of grouped values that aggregates the individual items of a more extensive table (such as from a database, spreadsheet, or business intelligence program) within one or more discrete categories. This summary might include sums, averages, or other statistics, which the pivot table groups together using a chosen aggregation function applied to the grouped values.

Although pivot table is a generic term, Microsoft trademarked PivotTable in the United States in 1994 (which was canceled in 2020).[1]

# Data Cleaning - Data Duplikat

Data cleaning / cleansing / wrangling adalah proses mendeteksi dan mengkoreksi (atau membuang) data - data yang tidak akurat atau corrupt dari suatu dataset. Data yang di-cleaning bisa saja karena data itu tidak lengkap, tidak benar, tidak akurat, ataupun tidak relevan dengan data yang ingin di-analisa. 

Setelah dilakukan data cleaning, dataset harus tetap konsisten sebelum ataupun setelah dilakukan pengkoreksian. Konsistensi sangat bergantung dengan bagaimana user melakukan metode pengkoreksian terhadap data yang di-cleaning. Bisa saja proses mem"bersih"kan data malah membuat data menjadi "kotor" (value dari dataset berubah). Data cleaning berbeda dengan data transformation.

Data cleaning terkadang hanya membersihkan data - data sederhana dikarenakan salah input (typo), nilai tidak ditemukan (null), ataupun nilai tidak valid (data outlier, overfit, stopword, dsb).

Tidak ada satu cara yang "pakem" dalam melakukan data cleaning, jadi silahkan bersih - bersih sesuai dengan kebutuhan analisa dataset anda.  

## `drop_duplicates`

### Sampel
Dataset kita buat duplikat dulu

```python
temp_df = pd.concat([movies_df,movies_df])

temp_df.shape
```

### Lakukan drop baris yang sama

```python
temp_df = temp_df.drop_duplicates() 
#akan mengeluarkan return value data diremove duplikatnya, lalu ditangkap temp_df

temp_df.shape
```

```python
temp_df.drop_duplicates(inplace=True)
```

Argument penting lain dalam drop_duplicates() adalah <code>keep</code> , yang bisa memiliki nilai:

    first: (default) Drop duplikat kecuali temuan pertama
    last: Drop duplikat kecuali temuan terakhir
    False: Drop semua isi, sehingga kosong

```python
temp_df2 = pd.concat([movies_df,movies_df])  # make a new copy

#temp_df2.drop_duplicates(inplace=True, keep=False)
temp_df2.drop_duplicates(inplace=True, keep=False)
temp_df2.shape
```

##  Menggunakan Fungsi

```python
# Membuat fungsi 
def rating_function(x):
    if x >= 8.0:
        return "good"
    else:
        return "bad"
```

## Mengaplikasikan fungsi

```python
movies_df["rating_category"] = movies_df["rating"].apply(rating_function)
#menambah kolom baru bernama "rating_category" dengan data rating yang dikenai fungsi

movies_df.head()
```

### Drop kolom

```python
del movies_df['rating_category']
movies_df.head()
```

### Membuat fungsi anonim menggunakan lambda

```python
movies_df["rating_category"] = movies_df["rating"].apply(lambda x: 'good' if x >= 8.0 else 'bad')

movies_df.head(2)
```
